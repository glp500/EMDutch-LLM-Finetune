```python
meteor = evaluate.load('meteor')
```

```python
# Create a dataframe for Mixtral-8x7B-Instruct scores
mixtral_bert_df = pd.DataFrame(columns=["Precision","Recall","F1"])
mixtral_meteor_df = pd.DataFrame(columns=["Value"])

for index,rows in text_df.iterrows():
  candidate = rows["Mixtral-8x7B-Instruct"]
  reference = rows['Translation']
  scorer = BERTScorer(model_type='bert-base-uncased')
  P, R, F1 = scorer.score([candidate], [reference])
  mixtral_bert_df.loc[index] = [P, R, F1]

  value = meteor.compute(predictions=[candidate], references=[reference])

  mixtral_meteor_df.loc[index] = [value]

```

```python
# Export files
mixtral_bert_df.to_csv('/content/drive/Othercomputers/My MacBook Pro/Thesis_Project/Data/translation_text_analysis/Bert Scores/mixtral_bert_scores.csv')
mixtral_meteor_df.to_csv('/content/drive/Othercomputers/My MacBook Pro/Thesis_Project/Data/translation_text_analysis/Meteor Scores/mixtral_meteor_scores.csv')

```
