```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/Volumes/Extreme SSD/Thesis_Project/Data/translation_test/test_text.csv')
```

```python
# Print number of rows in the dataframe
print(df.shape)
```

```python
df.head()
```

```python
# Make two charts for the two columns of the dataframe; Count the number of words in each row for the two columns
df['word_count_input'] = df['Input'].apply(lambda x: len(str(x).split()))
df['word_count_translation'] = df['Translation'].apply(lambda x: len(str(x).split()))
```

```python
# Plot the two charts with  word amount on the x-axis and the frequency on the y-axis
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(df['word_count_input'], bins=20, color='blue')
plt.title('Early Modern Dutch Texts')
plt.xlabel('Word count')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
sns.histplot(df['word_count_translation'], bins=20, color='red')
plt.title('English Translation Texts')
plt.xlabel('Word count')
plt.ylabel('Frequency')


# Change the x-labels to ""

# Create a large bolder title above the two charts named: "Word Count Distribution Per Sentence"
plt.suptitle('Word Count Distribution for each Sentence', fontsize=16)
plt.show()
```
