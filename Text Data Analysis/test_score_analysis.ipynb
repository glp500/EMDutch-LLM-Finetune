```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re

# Function to clean the cell data
def clean_tensor_data(cell):
    cell_str = str(cell)
    # Use regular expression to extract the value
    match = re.search(r'tensor\(\[(.*?)\]\)', cell_str)
    if match:
        return match.group(1)
    return cell

def clean_meteor_data(cell):
    cell_str = str(cell)  # Convert cell to string
    match = re.search(r"\{'meteor': (.*?)\}", cell_str)
    if match:
        return match.group(1)
    return cell
```

```python
# Importing the data
bert = pd.read_csv('/Volumes/Extreme SSD/Thesis_Project/Data/translation_text_analysis/Bert Scores/f1_scores.csv')
meteor = pd.read_csv('/Volumes/Extreme SSD/Thesis_Project/Data/translation_text_analysis/Meteor Scores/meteor_scores.csv')
```

```python
# Besides the input and translation columns, the data is in string format. We need to convert it to float
bert = bert.astype({'Mixtral-8x7B-Instruct': 'float', 'Llama-3-8B-Instruct': 'float', 'Aya-23-8B': 'float', 'Phi-3-medium-128k-instruct w/ Phi-3-mini-128k-instruct assistant model': 'float', 'ORPO-Llama-3-8b': 'float', 'Unsloth-Llama-3-8B-instruct': 'float'})
meteor = meteor.astype({'Mixtral-8x7B-Instruct': 'float', 'Llama-3-8B-Instruct': 'float', 'Aya-23-8B': 'float', 'Phi-3-medium-128k-instruct w/ Phi-3-mini-128k-instruct assistant model': 'float', 'ORPO-Llama-3-8b': 'float', 'Unsloth-Llama-3-8B-instruct': 'float'})
```

```python
# Get a summary of the data for each model and the meteor scores and f1 scores and create a new dataframe
summary_bert = pd.DataFrame(bert.describe())

summary_meteor = pd.DataFrame(meteor.describe())


```

```python
# Alter the style of the boxplots of bert and meteor scores, also shorten the names of the models to make them only the characters leading up to the first '-'
# # Each model should have its own own color: (Mixtral: red), (Llama 3: blue), (Aya: yellow), (Phi: green), (ORPO: orange), (Unsloth: purple)
# use the theme seaborn

```

```python
# Create two new dataframes that are the same as the orginal meteor and bert dataframes but sorted by words in each string in the 'Translation' column
bert_sorted = bert.sort_values(by='Translation'.split())
meteor_sorted = meteor.sort_values(by='Translation'.split())
```

```python
# Change the column names to be 'Mixtral', 'Llama', 'Aya', 'Phi', 'ORPO', 'Unsloth'
bert_sorted.columns = ['Input', 'Translation', 'Mixtral', 'Llama', 'Aya', 'Phi', 'ORPO', 'Unsloth']
meteor_sorted.columns = ['Input', 'Translation', 'Mixtral', 'Llama', 'Aya', 'Phi', 'ORPO', 'Unsloth']
```

```python
# Create two scatter plots that have the bert and meteor scores on the y-axis and the index of the dataframe on the x-axis
# On the x-axis make sure that the data is sorted by word count
plt.figure(figsize=(20, 10))
plt.scatter(bert_sorted.index, bert_sorted['Mixtral'], label='Mixtral')
plt.scatter(bert_sorted.index, bert_sorted['Llama'], label='Llama')
plt.scatter(bert_sorted.index, bert_sorted['Aya'], label='Aya')
plt.scatter(bert_sorted.index, bert_sorted['Phi'], label='Phi')
plt.scatter(bert_sorted.index, bert_sorted['ORPO'], label='ORPO')
plt.scatter(bert_sorted.index, bert_sorted['Unsloth'], label='Unsloth')
plt.title('Bert Scores')
plt.ylabel('Scores')
plt.xlabel('Index')
plt.legend()

# Show general 


plt.show()



```

```python
# For each model create a plot that has the index of the cell on the x-axis and the f1 score value on the y-axis. All the models should be on their own plot. Use bert_sorted as the data source
# On each of the plots also create an exponential trendline using the numpy polyfit function and the f1 scores, give this line a standard deviation
# Also create a horizontal line that represents the mean of the f1 scores for each model
# The plots should be in a 2x3 grid
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
models = ['Mixtral', 'Llama', 'Aya', 'Phi', 'ORPO', 'Unsloth']
for i in range(2):
    for j in range(3):
        model = models.pop(0)
        ax[i, j].scatter(bert_sorted.index, bert_sorted[model])
        ax[i, j].set_title(model)
        ax[i, j].set_xlabel('Index')
        ax[i, j].set_ylabel('F1 Score')
        ax[i, j].plot(bert_sorted.index, np.polyval(np.polyfit(bert_sorted.index, bert_sorted[model], 1), bert_sorted.index), label='Trendline', color='red')
        # Around the trendline create a shaded area that represents the standard deviation of the f1 scores
        ax[i, j].fill_between(bert_sorted.index, np.polyval(np.polyfit(bert_sorted.index, bert_sorted[model], 1), bert_sorted.index) - bert_sorted[model].std(), np.polyval(np.polyfit(bert_sorted.index, bert_sorted[model], 1), bert_sorted.index) + bert_sorted[model].std(), color='red', alpha=0.2)
        ax[i, j].axhline(bert_sorted[model].mean(), color='green', label='Mean')
        ax[i, j].legend()

        # Bold the title of the plot
        ax[i, j].title.set_fontweight('bold')
        ax[i, j].title.set_fontsize(12)

        # Bold the x-axis label and add increment markers by 10
        ax[i, j].xaxis.label.set_fontweight('bold')
        ax[i, j].xaxis.label.set_fontsize(12)
        ax[i, j].set_xticks(np.arange(0, len(bert_sorted.index), 10))
plt.tight_layout()

# Create a overarching title for the plots and give it a larger font size and make it bold
fig.suptitle('Bert Scores', fontsize=16, fontweight='bold')

# Increase the space between the supertitle and the plots
plt.subplots_adjust(top=0.9)

```

```python
# For each model create a plot that has the index of the cell on the x-axis and the meteor score value on the y-axis. All the models should be on their own plot. Use meteor_sorted as the data source
# On each of the plots also create an exponential trendline using the numpy polyfit function and the meteor scores, give this line a standard deviation
# Also create a horizontal line that represents the mean of the meteor scores for each model
# The plots should be in a 2x3 grid

fig, ax = plt.subplots(2, 3, figsize=(20, 10))
models = ['Mixtral', 'Llama', 'Aya', 'Phi', 'ORPO', 'Unsloth']
for i in range(2):
    for j in range(3):
        model = models.pop(0)
        ax[i, j].scatter(meteor_sorted.index, meteor_sorted[model])
        ax[i, j].set_title(model)
        ax[i, j].set_xlabel('Index')
        ax[i, j].set_ylabel('Meteor Score')
        ax[i, j].plot(meteor_sorted.index, np.polyval(np.polyfit(meteor_sorted.index, meteor_sorted[model], 1), meteor_sorted.index), label='Trendline', color='red')
        # Around the trendline create a shaded area that represents the standard deviation of the meteor scores
        ax[i, j].fill_between(meteor_sorted.index, np.polyval(np.polyfit(meteor_sorted.index, meteor_sorted[model], 1), meteor_sorted.index) - meteor_sorted[model].std(), np.polyval(np.polyfit(meteor_sorted.index, meteor_sorted[model], 1), meteor_sorted.index) + meteor_sorted[model].std(), color='red', alpha=0.2)
        ax[i, j].axhline(meteor_sorted[model].mean(), color='green', label='Mean')
        ax[i, j].legend()

        # Bold the title of the plot
        ax[i, j].title.set_fontweight('bold')
        ax[i, j].title.set_fontsize(12)

        # Bold the x-axis label and add increment markers by 10
        ax[i, j].xaxis.label.set_fontweight('bold')
        ax[i, j].xaxis.label.set_fontsize(12)
        ax[i, j].set_xticks(np.arange(0, len(meteor_sorted.index), 10))

plt.tight_layout()

# Create a overarching title for the plots and give it a larger font size and make it bold
fig.suptitle('Meteor Scores', fontsize=16, fontweight='bold')

# Increase the space between the supertitle and the plots
plt.subplots_adjust(top=0.9)

plt.show()
```
