```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


```

```python
#Load the data from the file
text_df = pd.read_csv("/Volumes/Extreme SSD/Thesis_Project/Data/translation_test/results_final.csv")
```

```python
text_df.head()
```

```python
# For every column besides text_df['Input'] & text_df[Translation] (The first two columns of the dataframe), remove these combination of characters: "['","["","']", and ""]" from every cell in the column
for column in text_df.columns[2:]:
    text_df[column] = text_df[column].str.replace("['","")
    text_df[column] = text_df[column].str.replace("[\"","")
    text_df[column] = text_df[column].str.replace("']","")
    text_df[column] = text_df[column].str.replace("\"]","")


```

```python
text_df.head()
```

```python
# Export the cleaned dataframe to a new csv file
text_df.to_csv("/Volumes/Extreme SSD/Thesis_Project/Data/translation_test/results_final_cleaned.csv", index=False)
```

```python
# Create a new dataframe that has the same shape as the original dataframe,however instead of displaying the text in the cells it will display the word count of the text in the cells
word_count_df = text_df.copy()

# For every column besides text_df['Input'] & text_df[Translation] (The first two columns of the dataframe), count the number of words in the cell and store the count in the cell
for column in word_count_df.columns:
    word_count_df[column] = word_count_df[column].str.split().str.len()
```

```python
word_count_df.head()
```

```python
# Export the word count dataframe to a new csv file
word_count_df.to_csv("/Volumes/Extreme SSD/Thesis_Project/Data/translation_test/results_final_word_count.csv", index=False)
```
